# Data Augmentation

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
base_dir = 'images'
train_dir = os.path.join(base_dir,'TRAIN')
test_dir = os.path.join(base_dir, 'TEST')
```

```python
print("Contenido de la carpeta TRAIN:")
print(os.listdir(train_dir))  # Esto debe mostrar las categorías dentro de TRAIN

print("Contenido de la carpeta TEST:")
print(os.listdir(test_dir))  # Esto debe mostrar las categorías dentro de TEST

# También puedes verificar las imágenes dentro de cada subcarpeta:
for category in os.listdir(train_dir):
    category_path = os.path.join(train_dir, category)
    if os.path.isdir(category_path):
        print(f"Imágenes en {category}:")
        print(os.listdir(category_path))  # Lista las imágenes dentro de cada categoría
```

```python
train_datagen = ImageDataGenerator(
							rescale = 1./255,
							# rotation_range = 10,
							# width_shift_range = 0.2,
						  #	height_shift_range = 0.2,
						  #	shear_range = 0.3,
							# zoom_range = 0.3,
							# horizontal_flip = True
       )
```

```python
train_generator = train_datagen.flow_from_directory(
							train_dir,
							target_size = (150, 150),
							batch_size = 1,
							class_mode ='categorical',
)

plt.figure()
#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(1, 5, figsize=(30, 8))

for i in range(5) :
  axarr[i].imshow(train_generator[0][0][0])
```

```python
train_generator = train_datagen.flow_from_directory(
							train_dir,
							target_size = (150, 150),
							batch_size = 64,
							class_mode ='categorical',
							)


images , labels = train_generator[0]

print(images.shape)
print("labels: " , labels)


plt.figure()
#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(1, images.shape[0], figsize=(30, 4))

for i in range(images.shape[0]) :
  axarr[i].imshow(images[i])
```

```python
print("Clases de imágenes en el dataset:")
print(train_generator.class_indices)  # Muestra las clases y sus índices

# Verifica los primeros 5 archivos que el generador carga
for i in range(5):
    image, label = next(train_generator)
    print(f"Imagen {i+1}: {train_generator.filenames[i]}")
    print(f"Clase {i+1}: {label}")

    print(f"Valores de los píxeles de la primera imagen:")
    print(images[0].min(), images[0].max())  # Mínimo y máximo valor de los píxeles
```

## Escalamiento de imagenes

```python
def scale_images(train_images, test_images):
  train_images = train_images/255.
  test_images = test_images/255.

  return (train_images, test_images)
```

```python
from tensorflow.keras import optimizers
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model = models.Sequential()
model.add(layers.Conv2D(10, (3, 3), activation="relu", input_shape = (150,150,3)))
model.add(layers.Flatten())
model.add(layers.Dense(256,activation='relu'))
model.add(layers.Dense(4,activation='sigmoid'))

model.summary()

model.compile(loss='categorical_crossentropy',
						optimizer=optimizers.RMSprop(learning_rate=2e-5),
						metrics=['acc'])
```

```python
history = model.fit(
						train_generator,
						epochs = 32)


acc = history.history['acc']
loss = history.history['loss']

epochs = range(1, len(acc)+1)

plt.plot(epochs,acc,'bo',label='train accuracy')
plt.title('train acc')
plt.legend()

plt.figure()

plt.plot(epochs,loss, 'bo', label ='training loss')
plt.title('train loss')
plt.legend()

plt.show()
```

```python
test_datagen = ImageDataGenerator(1./255)
test_generator = test_datagen.flow_from_directory(
					test_dir,
					target_size = (150, 150),
					batch_size = 20,
					class_mode= 'categorical')

test_loss, test_acc = model.evaluate(test_generator, steps = 25)
print('\ntest acc :\n', test_acc)
```
